1.1 Write a Python Program to implement your own myreduce() function which works exactly
like Python's built-in function reduce()

```
# from functools import reduce
# reduce(lambda a,b,c : a+b+c, [25, 25, 59,60])

def myreduce(operationFunction, l) :
  returnVal = l[0]
  for i in l[1:] :
    returnVal = operationFunction(returnVal, i)
  return returnVal
testList = [1,2,3,1.5,5,8,8.8]
print(myreduce(lambda a,b : b/a, testList))
print(myreduce(lambda a,b : b*a, testList))
print(myreduce(lambda a,b : a if(a > b) else b, testList))
```

1.2 Write a Python program to implement your own myfilter() function which works exactly
like Python's built-in function filter()

```
#from functools import filter
#print(list(filter(lambda x :x%2 ==0, [25, 25, 59,60])))
def myfilter(operationFunction, l) :
  returnList = []
  for i in l :
    if operationFunction(i) :
      returnList.append(i)
  return returnList
l = [25,50,67,25,94,125,120,147,22145,2364,5782]
print(list(myfilter(lambda x :x%2 ==0, l)))
print(list(myfilter(lambda x :x%5 ==0, l)))
print(list(myfilter(lambda x :x%3 ==0, l)))
print(list(myfilter(lambda x :x >= 100, l)))
```
